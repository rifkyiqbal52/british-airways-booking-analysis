```python
# Hitung distribusi booking_complete
status_counts = df['booking_complete'].value_counts()

# Buat bar chart
plt.figure(figsize=(6,4))
ax = sns.barplot(x=status_counts.index, y=status_counts.values, palette="Set2")

# Tambahkan label jumlah di atas bar
for i, v in enumerate(status_counts.values):
    plt.text(i, v + 100, str(v), ha='center', va='bottom', fontweight='bold')

plt.title("Distribusi Status Booking (Selesai vs Tidak Selesai)")
plt.xlabel("Booking Complete (0=Tidak, 1=Ya)")
plt.ylabel("Jumlah Booking")
plt.show()
```

```python
# Hitung jumlah booking per channel dan status
channel_summary = df.groupby(['sales_channel', 'booking_complete']).size().unstack(fill_value=0)

# Tambahkan kolom total dan conversion rate
channel_summary['Total'] = channel_summary[0] + channel_summary[1]
channel_summary['Conversion Rate (%)'] = round(channel_summary[1] / channel_summary['Total'] * 100, 2)

print(channel_summary)
```

```python
# Crosstab channel vs booking_complete
channel_success = pd.crosstab(df['sales_channel'], df['booking_complete'], normalize='index')

channel_success.plot(kind='bar', stacked=True, figsize=(8,5), colormap="Set2")
plt.title("Perbandingan Keberhasilan Booking per Sales Channel")
plt.ylabel("Proporsi")
plt.show()
```

```python
# Hitung jumlah booking per trip_type dan status
trip_summary = df.groupby(['trip_type', 'booking_complete']).size().unstack(fill_value=0)

# Tambahkan kolom total dan conversion rate
trip_summary['Total'] = trip_summary[0] + trip_summary[1]
trip_summary['Conversion Rate (%)'] = round(trip_summary[1] / trip_summary['Total'] * 100, 2)

print(trip_summary)
```

```python
trip_success = pd.crosstab(df['trip_type'], df['booking_complete'], normalize='index')

trip_success.plot(kind='bar', stacked=True, figsize=(8,5), colormap="Set3")
plt.title("Keberhasilan Booking per Trip Type")
plt.ylabel("Proporsi")
plt.show()
```

```python
# Hitung jumlah permintaan layanan tambahan
total = len(df)
extras = {
    "Extra Baggage": df['wants_extra_baggage'].sum(),
    "Preferred Seat": df['wants_preferred_seat'].sum(),
    "In-flight Meals": df['wants_in_flight_meals'].sum()
}

# Hitung persentase
for k, v in extras.items():
    percent = round(v / total * 100, 2)
    print(f"{k}: {v} booking ({percent}%)")
```

```python
extras = {
    "Extra Baggage": df['wants_extra_baggage'].sum(),
    "Preferred Seat": df['wants_preferred_seat'].sum(),
    "In-flight Meals": df['wants_in_flight_meals'].sum()
}

plt.figure(figsize=(6,4))
sns.barplot(x=list(extras.keys()), y=list(extras.values()), palette="Set1")
plt.title("Distribusi Permintaan Layanan Tambahan")
plt.ylabel("Jumlah Booking dengan Layanan Tambahan")
plt.show()
```

```python
top_routes = df['route'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_routes.values, y=top_routes.index, palette="Set2")
plt.title("Top 10 Rute dengan Jumlah Booking Terbanyak")
plt.xlabel("Jumlah Booking")
plt.ylabel("Rute")
plt.show()
```
